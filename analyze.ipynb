## Polars Descriptive Statistics Script
### Data Loading

```python
import polars as pl

marketing_data = pl.read_csv("mydata/ifood_df.csv")
```

### Data Manipulation

```python
marketing_data = marketing_data.with_columns(
    TotalSpending = 
    pl.col("MntWines")
    + pl.col("MntFruits")
    + pl.col("MntMeatProducts")
    + pl.col("MntFishProducts")
    + pl.col("MntSweetProducts")
)
marketing_data = marketing_data.with_columns(
    AcceptedOffer = 
    pl.col("AcceptedCmp1")
    + pl.col("AcceptedCmp2")
    + pl.col("AcceptedCmp3")
    + pl.col("AcceptedCmp4")
    + pl.col("AcceptedCmp5")
)
marketing_data = marketing_data.select([
    "Income",
    "TotalSpending",
    "AcceptedOffer",
    "MntWines",
    "MntFruits",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
])
```

### Data Visualization

```python
import seaborn as sns
sns.scatterplot(data=marketing_data, x="Income", y="TotalSpending", hue="AcceptedOffer")
```
